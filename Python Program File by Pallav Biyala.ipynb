#### Program-1: List Operations and Basic Statistics

```python
#Creating List containing first 20 natural numbers
L=[]
for i in range(1,21):
    L.append(i)
print("List:",L)

# Calculating sum of even numbers and product of odd numbers
even_sum=0
odd_prod=1
for i in L:
    if i%2==0:
        even_sum+=i
    else:
        odd_prod=odd_prod*i

print("The sum of all even numbers in list is:",even_sum)
print("The product of all odd numbers in list is:",odd_prod)

# Finding mean,median and standard deviation of list

# First Mean
s=0
for i in L:
    s+=i
mean=s/len(L)
print("Mean of the list is: ",mean)

# Second Median
L.sort()
if len(L)%2!=0: #There are odd number of elements
    median=L[int((len(L)-1)/2)]
else:
    median=(L[int(len(L)/2)]+L[int((len(L)/2) -1)])/2
print("Median of the list is:",median)
    
# Standard Deviation
dev=0
for i in L:
    dev=dev+(i-mean)**2
var=dev/len(L)
stdev=(var)**0.5
print("Standard deviation of the list is:",stdev)
```

#### Program-2: Mutable vs. Immutable Data Types

#### Sol.2: The difference between mutable and immutable data types is that we can change as well as edit the mutable datatypes like list, dictionary etc. while we can't edit the immutable datatypes like string, tuple etc. If we try to change immutable datatypes then the python will raise an error. For example: 

```python
#List are mutable:
L=[1,2,3,4,5,6,7]
L[2]=13821
print(L)
```

```python
# Strings ar immutable:
s="My name iz Pallav"
s[3]="a"
print(s)
```

#### Hence we can edit modfiy mutable datatypes but can't modfiy immutable one. 

#### Program-3: Fibonacci Sequence 

```python
# Creating a list which will contain the fibonacci sequence
L=[]
n=int(input("Enter the nth term of fibonacci sequence: "))

# Fibonacci --> Sum of previous two terms will be next term
s=0
for i in range(n+1):
    if i==0:
        continue
    elif i==1 or i==2:
        s=1
        L.append(s)
    else:
        s=L[-1]+L[-2]
        L.append(s)
print("The list of fibonacci sequence till",n,"terms is: ",L)

# Plotting a linear graph using matplotlib library
import matplotlib.pyplot as graph
graph.plot(L)
graph.show()
```

#### Program-14: Word Filtering and Sorting

```python
# Taking input from user 
s=str(input("Enter a string: "))

# Splitting the string into individual words
L=s.split()
print("The list of words is: ",L)

M=[]
# Removing all the words less than 4 alphabets
for i in L:
    if len(i)<4:
        continue
    else:
        M.append(i)
print("The list after removing all the characters with length less than 4 is: ",M)

# Sorting the remianing words alphabetically
for i in range(len(M)): 
    M[i]=M[i].lower()        # So that the python doesn't keep small letters before capital 
M.sort()
print("The list after sorting the words is: ",M)

# Plotting the bar graph for word lengths and their frequency 
import matplotlib.pyplot as graph

C=s.split()
D=[]
for i in C:
    D.append(len(i))        # D contains length of all characters

# Creating a dictionary C which will contain word length as key and their frequency as values
C={}
for i in D:
    if i not in C:
        C[i]=1
    else:
        C[i]+=1     # C now contains a dictionary with frequency of each character

print("The dictionary with word length as character and frequency as value is: ",C)
        
E=[]
F=[]
for i in C:
    E.append(C[i])   # E contains all the values (frequency)
    F.append(i)      # F contains all the word lengths

graph.bar(F,E,color="cyan",edgecolor="black")
graph.xlabel("Word Lenghts")
graph.ylabel("Frequency")
graph.show()
```

#### Program-22: Normalization

```python
# Normalization
# Normalized value = (current value - min(list))/(max(list)-min(list))


# Creating a list of integers given by user
n=input("Enter a list of numbers: (seperated by space) ").split()
L=[]
for i in n:
    L.append(int(i))

# Creating a list which will contain all the normlaized values    
N=[]
for i in L:
    a=round((i-min(L))/(max(L)-min(L)),3)   # a is normalized value of i in L
    N.append(a)
print("The normalized list of numbers is: ",N)

# Plotting a line plot
import matplotlib.pyplot as graph
graph.plot(N)
graph.xlabel("Index")
graph.ylabel("Normalized Value")
graph.show()
```
